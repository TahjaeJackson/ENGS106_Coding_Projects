### Training Data Generation 

For this assignment, I generated the training data by personally playing rock–paper–scissors against the computer for 100 rounds and manually recording the outcomes. This approach ensured that the data reflected a consistent decision-making style and a single set of behavioral patterns.

If multiple people each generated a small amount of training data, the resulting dataset would likely contain a mixture of different strategies, habits, and biases. While this could increase diversity in the data, it may also make it harder for a simple model to learn clear patterns, especially with a limited number of total rounds. Conflicting tendencies across players could introduce noise and reduce the effectiveness of pattern exploitation.


```python
from collections import Counter

np.random.seed(0)
random.seed(0)

history = np.loadtxt(r"./training.txt", dtype=int).reshape(-1,3)

print(history)
#Define your algorithm here or define your own functions and procedure

# create a data structure to store the best play to beat a tool
# 0 - rock, 1-paper, 2 - scissors
best_play = {0:1, 1:2, 2:0}
prob_given = {0:{0:0,1:0,2:0},1:{0:0,1:0,2:0},2:{0:0,1:0,2:0}}
def rock_paper_scissor(history=None, last_choice=None):
    # define your function here
    
    # if there is no history or only two plays stored, randomly choose a tool 
    if history is None or len(history) < 2:
        return random.randint(0,2)
    
    # training based on history
    
    # extract the CPU's choices [cpu, your move, score]
    cpu = []
    for i in range(len(history) - 1):
        current_tool = history[i][0]
        next_tool = history[i + 1][0]

        prob_given[current_tool][next_tool] += 1

    
    # Look at the opponent's most recent move
    if last_choice is None:
        last_tool = history[-1][0]
    else:
        last_tool = last_choice

    # Predict the opponent's next move (most frequent transition)
    predicted_next = max(
        prob_given[last_tool],
        key=prob_given[last_tool].get
    )
    print(prob_given)
    return best_play[predicted_next]
```

```python
print(f'Algorithm says you should choose: {rock_paper_scissor(history)}')
```

# Testing the model

```python
move_names = {0: "rock", 1: "paper", 2: "scissors"}

evaluation_data = []
wins = losses = ties = 0

last_cpu_move = None  # track opponent's most recent move

for i in range(100):
    print(f"\nRound {i+1}")

    model_move = rock_paper_scissor(history, last_choice=last_cpu_move)
    print("Model recommends:", move_names[model_move])

    cpu_input = input("Enter computer move (rock/paper/scissors): ").strip().lower()
    cpu_move = {"rock": 0, "paper": 1, "scissors": 2}[cpu_input]
    
    # determining the outcome of the round
    if model_move == cpu_move:
        score = 0
        ties += 1
    elif (model_move - cpu_move) % 3 == 1:
        score = 1
        wins += 1
    else:
        score = -1
        losses += 1

    evaluation_data.append([cpu_move, model_move, score])

    # update history and last move
    history = np.vstack([history, [cpu_move, model_move, score]])
    last_cpu_move = cpu_move
```

```python


np.savetxt("evaluation.txt", evaluation_data, fmt="%d")
print(f"Wins: {wins}, Losses: {losses}, Ties: {ties}")
```
